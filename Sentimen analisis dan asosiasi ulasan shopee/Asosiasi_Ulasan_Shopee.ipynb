```python
transactions = [
    ["apple", "banana", "cherry"],
    ["banana", "grape"],
    ["apple", "banana", "cherry", "grape"],
    ["cherry", "grape"],
    ["apple", "cherry"]
]

# Menerapkan FP-Growth untuk menemukan itemset yang sering muncul
patterns = pyfpgrowth.find_frequent_patterns(transactions,2)

# Menampilkan itemset yang sering muncul
print("Frequent Itemsets:")
for itemset, support in patterns.items():
    print(f"{itemset}: {support}")

# Menerapkan FP-Growth untuk menemukan association rules
rules = pyfpgrowth.generate_association_rules(patterns, 0.5)

# Menampilkan association rules
print("\nAssociation Rules:")
for rule, confidence in rules.items():
    print(f"Rule: {rule[0]} -> {rule[1]}, Confidence: {confidence}")
```
